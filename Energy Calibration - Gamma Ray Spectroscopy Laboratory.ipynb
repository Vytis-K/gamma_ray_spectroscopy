```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of values for E
E = np.linspace(0.01, 2.0, 100)  # You can adjust the range and number of points as needed

# Calculate Delta E / E using the equation
delta_E_over_E = np.sqrt((0.14 / np.sqrt(E))**2 + 0.41**2)

# Calculate 1 / E
one_over_E = 1 / E

plt.figure(figsize=(8, 6))
plt.plot(one_over_E, delta_E_over_E, label=r'$\frac{\Delta E}{E}$ vs $\frac{1}{E}$', color='b')
plt.xlabel(r'$\frac{1}{E}$')
plt.ylabel(r'$\frac{\Delta E}{E}$')
plt.title(r'$\frac{\Delta E}{E}$ vs $\frac{1}{E}$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck constant in Js
m_e = 9.109e-31  # Mass of electron in kg
c = 3.0e8  # Speed of light in m/s
initial_energy_keV = 0.662  # Initial energy in keV
initial_wavelength = (h * c) / (initial_energy_keV * 1e3)  # Initial wavelength in meters

# Scattering angles from 0 to 180 degrees
theta_degrees = np.linspace(0, 180, 181)
theta_radians = np.deg2rad(theta_degrees)

# Calculate change in wavelength for each angle
delta_lambda = (h / (m_e * c)) * (1 - np.cos(theta_radians))

# Calculate change in energy using the wavelength change
delta_energy_keV = (h * c) / (delta_lambda * 1e3) - initial_energy_keV

plt.figure(figsize=(8, 6))
plt.plot(theta_degrees, delta_energy_keV)
plt.title('Compton Scattering of 0.662 keV Gamma Ray')
plt.xlabel('Scattering Angle (degrees)')
plt.ylabel('Change in Energy (keV)')
plt.grid(True)
plt.show()
```
